# 04 — Explainability
IEEE-CIS Fraud Detection

```python
import pandas as pd
import xgboost as xgb
import shap
import matplotlib.pyplot as plt

# Load processed features and trained model
df = pd.read_parquet('../data/processed/features.parquet')
X  = df.drop('isFraud', axis=1)
y  = df['isFraud']

model = xgb.XGBClassifier()
model.load_model('../models/xgb_fraud.bin')
```

```python
# SHAP values on a sample of 500 rows (full dataset is slow)
sample = X.sample(500, random_state=42)
explainer  = shap.TreeExplainer(model)
shap_values = explainer.shap_values(sample)
```

```python
# Summary plot — top 20 most important features
shap.summary_plot(shap_values, sample, max_display=20)
```

```python
# Waterfall plot — explain a single fraud transaction
fraud_sample = X[y == 1].sample(1, random_state=42)
shap_single  = explainer(fraud_sample)
shap.plots.waterfall(shap_single[0])
```

```python
# Bar plot — mean absolute SHAP values
shap.summary_plot(shap_values, sample, plot_type='bar', max_display=20)
```
